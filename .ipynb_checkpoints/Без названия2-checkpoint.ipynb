```python
import os
import requests
import django
os.environ.setdefault('DJANGO_SETTINGS_MODULE', 'blog.settings')
os.environ.setdefault('DJANGO_ALLOW_ASYNC_UNSAFE', 'true')
django.setup()
```

```python
from django.contrib.auth import get_user_model
User = get_user_model()
user = User.objects.get(id=1)
```

```python
from article.models import Post, Comment
Post.objects.all()
```

```python
url = "https://jsonplaceholder.typicode.com/posts"
response = requests.get(url)
data = response.json()
```

```python
data[0]
```

```python
Post.objects.all().delete()
post_cid_to_id_dict = {}
for post in data:
    p = Post.objects.create(author=user, slug=post['id'], title=post['title'], body=post['body'])
    post_cid_to_id_dict[post['id']] = p.id
```

```python
url = "https://jsonplaceholder.typicode.com/comments"
response = requests.get(url)
data = response.json()

data[0]
```

```python
comments_list = []

Comment.objects.all().delete()
for comment in data:
    #post = Post.objects.get(id=post_cid_to_id_dict[comment['postId']])
    comments_list.append(Comment(
                                post_id=post_cid_to_id_dict[comment['postId']], 
                                author=comment['email'], 
                                body=comment['body']
                                ))
    #Comment.objects.create(
    #    post_id=post_cid_to_id_dict[comment['postId']], 
    #    author=comment['email'], 
    #    body=comment['body']
    #)
    
Comment.objects.bulk_create(comments_list)
```
